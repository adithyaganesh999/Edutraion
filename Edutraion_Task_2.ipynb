# Task - 2 Movie Rating Prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, r2_score

# Step 1: Load dataset
df = pd.read_csv("/content/IMDb Movies India.csv", encoding="ISO-8859-1")

# Step 2: Clean data - remove rows without ratings
df = df.dropna(subset=['Rating'])
```

```python
df.info()
```

```python
# Step 3: Fill missing values with defaults
df['Genre'] = df['Genre'].fillna('Unknown')
df['Director'] = df['Director'].fillna('Unknown')
df['Actor 1'] = df['Actor 1'].fillna('Unknown')
df['Actor 2'] = df['Actor 2'].fillna('Unknown')
df['Actor 3'] = df['Actor 3'].fillna('Unknown')
df['Duration'] = df['Duration'].fillna('0 min')
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(int)

# Step 4: Encode text columns
encoder = LabelEncoder()
df['Genre'] = encoder.fit_transform(df['Genre'])
df['Director'] = encoder.fit_transform(df['Director'])
df['Actor 1'] = encoder.fit_transform(df['Actor 1'])
df['Actor 2'] = encoder.fit_transform(df['Actor 2'])
df['Actor 3'] = encoder.fit_transform(df['Actor 3'])
```

```python
# Step 5: Prepare features and target
X = df[['Duration', 'Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']]
y = df['Rating']

# Step 6: Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Step 7: Train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Step 8: Predict and evaluate
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("Mean Absolute Error:", mae)
print("R-squared Score:", r2)
```
